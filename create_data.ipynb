```python
from faker import Faker
import pymongo
import pandas as pd
import random

fake = Faker()

client = pymongo.MongoClient("mongodb://localhost:27017/")

mydb = client["Vulns"]

teams_col = mydb["teams"]
assets_col = mydb["assets"]
```

# Assets

```python
teams_countries = []

for i in list(teams_col.find({}, { "name": 1 ,"_id":0})):
    teams_countries.append(i["name"])

teams_countries
```

```python
from faker.providers import internet
import uuid
fake = Faker()
fake.add_provider(internet)


seen = []
assets = []

for i in range(20_000):

    hostname = str(uuid.uuid4())

    
    while True:
        ip = fake.ipv4_private()
        
        if ip not in seen:
            seen.append(ip)
            break
        

    
    n = random.randint(1,5)
    rand_teams = random.sample(teams_countries, n)


    asset = {
        "hostname": hostname,
        "ip": ip,
        "team": rand_teams ,
        "ACR": round(random.uniform(0,10), 1)
    }
    
    assets.append(asset)
```

```python
assets_col.insert_many(assets)
```

# Assets Neo4j

```python
from neo4j import GraphDatabase

uri = "bolt://localhost:7687"  
username = "neo4j"  
password = "" 

driver = GraphDatabase.driver(uri, auth=(username, password))
```

```python
driver.verify_connectivity()
```

```python
def insert_nodes(nodes):
    with driver.session() as session:
        for node in nodes:
            query = """
            CREATE (n:Asset {
                hostname: $hostname,
                ip: $ip,
                ACR: $acr
            })
            WITH n
            UNWIND $team AS team
            MERGE (t:Team {name: team})
            MERGE (n)-[:Pertence]->(t)
            """
            session.run(query, hostname=node['hostname'], ip=node['ip'], acr=node['ACR'], team=node['team'])
```

```python
all_assets =list(assets_col.find(projection={"_id":False}))

insert_nodes(all_assets)
```

```python

for asset in all_assets:
    

    n = random.randint(1,10)


    rand_conections = random.sample(all_assets, n)

    for con in rand_conections:
        
        with driver.session() as session:
            query = """
            MATCH (a1:Asset {hostname: $hostname1}), (a2:Asset {hostname: $hostname2})
            CREATE (a1)-[:conectado_com]->(a2)
            """
            session.run(query, hostname1=asset["hostname"], hostname2=con["hostname"])
```

# Exports

```python
cve = pd.read_csv("cve.csv").rename(columns={"Unnamed: 0": "cve"}).to_dict("records")

for vuln in cve:

    cvss = vuln["cvss"]
    
    severity = ""
    if cvss < 2:     
        severity = "Info"
    
    elif cvss <= 2 and cvss > 5:
        severity = "Low"
    
    elif cvss <= 5 and cvss > 7:
        severity = "Medium"
    
    elif cvss <= 7 and cvss > 8.5:
        severity = "High"
    
    elif cvss <= 8.5:
        severity = "Critical"


    vuln.update({"severity": severity})
```
